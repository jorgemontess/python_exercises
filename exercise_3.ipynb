Ejercicio: Análisis de Datos de Ventas Trimestrales

Contexto: Imagina que trabajas para una empresa de ventas al por menor y te han proporcionado un conjunto de datos que contiene información sobre las ventas trimestrales en diferentes regiones geográficas durante el último año. La empresa está interesada en comprender las tendencias de ventas trimestrales y comparar el rendimiento entre regiones.

Problema: Utiliza Pandas para cargar los datos y realizar un análisis básico que responda a las siguientes preguntas:

¿Cuál es el trimestre con las ventas más altas en cada región? --> Trimestre 4

¿Cuál es el trimestre con las ventas más bajas en cada región? --> Trimestre 1

¿Cuál es el trimestre con el mayor crecimiento de ventas en cada región (en comparación con el trimestre anterior)? --> Trimestre 2

```python
import pandas as pd
```

```python
datos_ventas_trimestrales = {
    'Region': ['Norte', 'Sur', 'Este', 'Oeste'],
    'Trimestre1': [10000, 12000, 8000, 15000],
    'Trimestre2': [11000, 13000, 8200, 15500],
    'Trimestre3': [12000, 13500, 8400, 16000],
    'Trimestre4': [12500, 14000, 8500, 16500]
}
```

```python
df_ventas_trimestrales = pd.DataFrame(datos_ventas_trimestrales)
df_ventas_trimestrales
```

```python
trimestre_maximo = df_ventas_trimestrales[['Trimestre1', 'Trimestre2', 'Trimestre3', 'Trimestre4']].idxmax(axis=1)
df_ventas_trimestrales['Trimestre Maximo'] = trimestre_maximo
df_ventas_trimestrales
```

```python
#Eligiendo el valor minimo, axis = 1 agarra los valores del dataframe en columnas, no en filas
trimestre_minimo = df_ventas_trimestrales[['Trimestre1', 'Trimestre2', 'Trimestre3', 'Trimestre4']].idxmin(axis=1)

# Agregando los datos a un nuevo campo Trimestre Minimo en nuestro DataFrame
df_ventas_trimestrales['Trimestre Minimo'] = trimestre_minimo
df_ventas_trimestrales
```

```python
# Calculando la diferencia trimestral, .sub sirve para restar
diferencia_trimestral = df_ventas_trimestrales[['Trimestre2', 'Trimestre3', 'Trimestre4']].sub(df_ventas_trimestrales[['Trimestre1', 'Trimestre2', 'Trimestre3']].values)
#diferencia_trimestral
# Eligiendo el valor mayor de la tabla de diferencia_trimestral
maximo_crecimiento_trimestral = diferencia_trimestral.idxmax(axis=1)

# Agregando los datos a un nuevo campo Maximo_Crecimiento_Trim en nuestro DataFrame
df_ventas_trimestrales['Maximo_Crecimiento_Trim'] = maximo_crecimiento_trimestral


df_ventas_trimestrales
```
