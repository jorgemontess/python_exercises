Contexto: Imagina que trabajas para una empresa de ventas al por menor y te han proporcionado un conjunto de datos que contiene información sobre las ventas mensuales en diferentes categorías de productos durante el último año. La empresa está interesada en visualizar las tendencias de ventas a lo largo del año para cada categoría de productos.

Problema: Utiliza Pandas para cargar los datos y Matplotlib para crear un gráfico de líneas que muestre las ventas mensuales en cada categoría de productos a lo largo del año. Asegúrate de agregar etiquetas adecuadas a los ejes, un título informativo y una leyenda para las diferentes categorías de productos.

Una vez que hayas creado el gráfico, comenta sobre cualquier patrón o tendencia interesante que observes en los datos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Nuestro DataSet en forma de diccionario
datos_ventas = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'],
    'Electrónicos': [10000, 12000, 13000, 11000, 14000, 15000, 16000, 17000, 18000, 19000, 20000, 21000],
    'Ropa': [8000, 8500, 9000, 9500, 10000, 10500, 11000, 11500, 12000, 12500, 13000, 13500],
    'Hogar': [6000, 6500, 7000, 7500, 8000, 8500, 9000, 9500, 10000, 10500, 11000, 11500]
}
```

```python
# Convertimos nuestro dataset a un DataFrame para poder manipular y graficar los datos
df_ventas = pd.DataFrame(datos_ventas)
df_ventas
```

```python
# Creamos nuestros axis(ejes) x, y con la libreria numpy, los convertimos en np.array
x_meses = np.array(df_ventas['Mes'])
y_electro = np.array(df_ventas['Electrónicos'])
y_ropa = np.array(df_ventas['Ropa'])
y_hogar = np.array(df_ventas['Hogar'])
```

```python
# Dibujando nuestro lienzo, figuras de 35 x 5 y sin compartir el axis y
fig, axs =  plt.subplots(1, 3, figsize=(35, 5), sharey=False)

#Declarando en axs[n] a que figura le queremos hacer el cambio

# Figura 1
axs[0].bar(x_meses, y_electro)
axs[0].set_title('Electrodomestico')

# Figura 2
axs[1].bar(x_meses, y_ropa)
axs[1].set_title('Ropa')        # Declaramos el titulo de la figura 2

# Figura 3
axs[2].bar(x_meses, y_hogar)
axs[2].set_title('Hogar')


# Declaramos el titulo de todo el lienzo
fig.suptitle('Productos')
```

```python
fig, ax = plt.subplots(figsize = (15, 10))
ax.plot(x_meses, y_electro, label = 'Electrodomesticos', marker = 'o')
ax.plot(x_meses, y_ropa, label = 'Ropa', marker = 'o')
ax.plot(x_meses, y_hogar, label = 'Hogar', marker = 'o')

ax.legend()  # Cuadro que nos muestra el color = titulo de las lineas (datos)
plt.show()
```
